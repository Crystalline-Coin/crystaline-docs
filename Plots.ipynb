```python
import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d
import numpy as np
```

Here we set the resolution value of plots

```python
resolution_value = 300
```

activity.tex - TimeFactor plot (h_t)



```python
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

temp1 = 10
temp2 = 10
def h(x):
   return temp1/(x + temp2)

x = np.linspace(-10, 100, 500)

plt.plot(x, h(x), color='black')
#plt.title()
plt.xlabel('Elapsed time')                           # label of x axis
plt.ylabel('timeFactor')                             # label of y axis
plt.xticks([0])                                      # x axis range
plt.yticks([1])                                      # y axis range
plt.xlim([0,100])                                     # x axis label numbers
plt.ylim([0,2])                                    # y axis label numbers
plt.gcf()
plt.savefig("h_t", facecolor='w', bbox_inches="tight",
            pad_inches=0.3, transparent=True , dpi=resolution_value)
plt.show()

```

activity.tex - SizeFactor plot (k_s)


```python
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

T = 120
beta = 75

def f(x):
   return -(T/(x + T/beta)) + beta

def g(x):
   return beta + x - x

x = np.linspace(0 , 25 , 200)

plt.plot(x, f(x), color='black')
plt.plot(x , g(x) , color = 'darkblue' , linestyle='dashed')
plt.xlabel('size')                           # label of x axis
plt.ylabel('sizeFactor')                     # label of y axis
plt.xlim([0,25])                             # x axis range
plt.ylim([0,100])                            # y axis range
plt.xticks([0])                              # x axis label numbers
plt.yticks([0])                              # y axis label numbers
plt.gcf()
plt.savefig("k_s", facecolor='w', bbox_inches="tight",
            pad_inches=0.3, transparent=True , dpi=resolution_value)
plt.show()
```

Meging above plots into one 3d plot 

```python
from matplotlib import cm

fig = plt.figure()
ax = plt.axes(projection='3d')

xline = np.linspace(0, 150, 100)
yline = np.linspace(0, 20, 100)

xline, yline = np.meshgrid(xline, yline)
def zi(x, y):
    return h(x) * f(y)

zline = zi(xline , yline)

plt.xticks([])                              # x axis label numbers
plt.yticks([])                              # y axis label numbers


ax.plot_surface(xline, yline, zline , rstride=5, cstride=5 , cmap='plasma' , edgecolor='none')                   # more colors can be used using cm.(...) in cmap argument
ax.set_xlabel('timeFactor')
ax.set_ylabel('sizeFactor')
ax.set_zlabel('activity')

figure = plt.gcf()
plt.tight_layout()
figure.set_size_inches(8, 6)
plt.savefig("Activity_plot.png", format="png", dpi=resolution_value)
```
